```python
#Функция расчета движения основной суммы по сделке 
def CF_Amount(type_s, par, col):
    
    #Исходные параметры сделки
    type_sd = par['type_sd']
    port = par['ind_portf']
    d_beg = par['data_begin']
    d_end = par['data_end']
    sh_1 = par['scheme_sd_1']
    amount = par['amount']
    defflag = par['defflg']
    defdate = pd.to_datetime (par['defdate'])
    lgd = par['LGD']
    
    #print(par['id_sd'])
    #stop = (input('нажмите для продолжения - вход в CF_Amount'))
    
    v1='Вложения в ценные бумаги и прочие финансовые активы, оцениваемые через ПУ'
    v2='Вложения в ценные бумаги и прочие финансовые активы, оцениваемые через ПСД'
    v3='Вложения в ценные бумаги и прочие финансовые активы, оцениваемые по АС'
    
    if port!='Операция' or type_sd == v1 or type_sd == v2 or type_sd == v3:
        raise ValueError('Incorrect port')
    else:
        if sh_1=='БезСр': # CF по бессрочным сделкам (УК, ОС и НА и т.п.)
            date_backets = pd.date_range(start = d_beg, periods=1)
            #print('Без срока', date_backets)
            df = pd.DataFrame(columns=col,index=date_backets).fillna(0)
            df.iloc[0,13] = amount
            df.iloc[0,14] = - amount
            return(df)
        elif sh_1== 'До востр' or sh_1=='Ежедн':
            date_end = d_beg+datetime.timedelta(days=1)
        else:
            date_end = d_end
     
        #Сдвиг графика на даты денежного потока по отношению к датам начисления
        delta_beg = pd.to_datetime(d_beg)
        delta=int(delta_beg.day)
        
        #Даты промежуточных выплат основного долга
        if sh_1=='Еженед':
            date_gap ='W'  
        elif sh_1=='Ежем':
            date_gap ='M' 
        elif sh_1=='Ежекв':
            date_gap ='Q'
        elif sh_1=='Раз в 6М':
            date_gap ='6M'
        elif sh_1=='Ежег':
            date_gap ='A' 
        else:
            date_gap ='NoPr'
    
        date_backets1 = pd.date_range(start = d_beg, end = date_end, periods=2)
        df = pd.DataFrame(columns=col,index=date_backets1).fillna(0)
        #print('Дата начала и дата окончания', date_backets1)
        #print('sh_1', sh_1)
        #print('date_gap', date_gap)
        
        if date_gap == 'NoPr': # CF по сделкам с погашением одной суммой
            df.iloc[0,13] = amount
            df.iloc[0,14] = - amount
            df.iloc[1,13] = - amount
            df.iloc[1,14] = amount
            #print('CF_Amount - шаг 11', df)
            #stop = (input('нажмите для продолжения - CF_Amount шаг 11 - поток на data_end'))
            
        else: # CF по сделкам с равномерным погашением суммы требования/обязательства
            date_backets2 = pd.date_range(d_beg, d_end, freq = date_gap)+datetime.timedelta(days=delta) 
            #print('Промежуточные даты гашения долга', date_backets2)
            df1 = pd.DataFrame(columns=col,index=date_backets2).fillna(0)
            df = pd.concat([df, df1]) 
            df = df.sort_index()
            dfw = df.groupby(df.index).sum()
            #print(type(dfw))
            #stop = (input('нажмите для продолжения - агрегация индексов'))
            df = dfw
            df.iloc[0,13] = amount
            df.iloc[0,14] = - amount
            n=len(date_backets2)
            #print(n,amount/n)
            #stop = (input('нажмите для продолжения - CF_Amount шаг 11 - поток на data_end'))
            ch_sum=-amount/n
            for i in range(1,n):
                df.iloc[i,13]=ch_sum
                df.iloc[i,14]=df.iloc[i-1,14]-df.iloc[i,13]
        #Модификация графиков потоков при дефолте или реструктуризации
        if defflag == 'Дефолт':
            #print('4. до модификации', df.index)
            df=df.loc[(df.index<=defdate)]
            #print('4. отрезали хвост', df.index)
            date_end_def = defdate.replace(defdate.year + 1)
            date_backets3 = pd.date_range(start = date_end_def, periods=1)
            df2 = pd.DataFrame(columns=col,index=date_backets3).fillna(0)
            df = pd.concat([df, df2])
            #print('4. добавили дату реализации залога', df.index)
            df = df.sort_index()
            dfw = df.groupby(df.index).sum()
            df = dfw
            df.iloc[-1,13] = - df.iloc[-2,14]*lgd
            df.iloc[-1,14] = df.iloc[-2,14] + df.iloc[-1,13]
            #print("4. возврат после дефолта", df.iloc[-1,13])
            #print("4. график после дефолта", df.index)
        elif defflag =='Реструкт':
            if(defdate>=d_end):
                raise ValueError('Incorrect defdate')
            else:
                #print('5. до модификации', df.index)
                dfbeg=df.loc[(df.index<=defdate)]
                #print('5. график до реструктуризации', dfbeg.index)
                dffin = df.loc[(df.index>defdate)]
                #print('5. график после реструктуризации до сдвига', dffin.index)
                dffin1_ind = dffin.index + datetime.timedelta(days=365)
                #print('индекс после реструктуризации после сдвига', dffin1_ind)
                dffin = dffin.set_index(dffin1_ind)
                #print('5. индекс после реструктуризации после сдвига', dffin.index)
                df = pd.concat([dfbeg, dffin])
                df = df.sort_index()
                dfw = df.groupby(df.index).sum()
                df = dfw
                #print("5. график после реструктуризации", df.index)
                #print("5. ОСЗ после реструктуризации", df.change_acc_cur)
        else:
            print("no def")
        #print(df.change_acc_cur)
        #print(df.acc_cur)
        #stop = (input('нажмите для продолжения - конец CF_Amount'))
    return(df)
```
